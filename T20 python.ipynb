```python
#import necessary libraries
import pandas as pd
import json
```

### match_summary

```python
data = open('t20_wc_match_results.json')
match = json.load(data)
match

match = pd.DataFrame(match[0]['matchSummary'])
match.tail(11)
```

```python
match.rename({'scorecard': 'id'}, axis = 1, inplace = True)
match.tail()
```

```python
match.to_csv('Downloads/t_20/t20_json_files/match_results.csv', index = False)
```

```python
match_dict = {}

for index, row in match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_dict[key1] = row['id']
    match_dict[key2] = row['id']
    
match_dict
```

### Bowling_summary

```python
data = open('t20_wc_bowling_summary.json')
bowling = json.load(data)
records = []
for rec in bowling:
    records.extend(rec['bowlingSummary'])
records[:3]
```

```python
bowling = pd.DataFrame(records)
print(bowling.shape)
bowling.tail(11)
```

```python
bowling['id'] = bowling['match'].map(match_dict)
bowling.tail(11)
```

```python
bowling.to_csv('Downloads/t_20/t20_json_files/t20_json_files/new_bowling_summary.csv', index = False)
```

### Player_info

```python
with open('Downloads/t_20/t20_json_files/t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
players = pd.DataFrame(data)
players.tail(10)
```

```python
players.shape
```

```python
players['name'] = players['name'].apply(lambda x: x.replace('â€', ''))
players['name'] = players['name'].apply(lambda x: x.replace('†', ''))
players['name'] = players['name'].apply(lambda x: x.replace('\xa0', ''))
players.tail(11)
```

```python
players[players['team'] == 'India']
```

```python
players.to_csv('Downloads/t_20/t20_json_files/t20_json_files/new_player_info.csv', index = False)
```
